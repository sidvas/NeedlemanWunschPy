```python
def needleman_wunsch(seq1, seq2, scoring_matrix, gapPen):
    '''function to perform needleman_wunsch algorithm
    - given two genetic sequences, a 4x4 scoring_matrix and the penalty for gap
    , the function generates the dp matrix and runs the traceback
    to return the score and 1 of the optimal alignments'''
    score_matrix = np.zeros((len(seq1)+1, len(seq2)+1)) #declare empty dp matrix of zeroes
    parent_matrix = np.zeros((len(seq1)+1, len(seq2)+1), dtype=int) #matrix to keep track of tracebacks
    diag = 3 #set-default values to distinguish between where score came from
    left = 2
    up = 1
    for i in range(1, len(seq1)+1): #initialize first column of scoring matrix with the penalty gaps
        score_matrix[i][0] = gapPen*i
    for j in range (1, len(seq2)+1): #initialize first row of scoring matrix with the penalty gaps
        score_matrix[0][j] = gapPen *j
    #print score_matrix
    parent_matrix[0,1:] = left #set parents for first row and column
    parent_matrix[1:,0] = up
    #print parent_matrix
    for i in range(1, len(seq1)+1): #building dp matrix and arrows
        for j in range(1, len(seq2)+1):
            ind1 = 0 #indices to determine match/mismatch score from the scoring_matrix
            ind2 = 0
            
            if seq1[i-1] == 'A': #determine which element from sequence 1 we are looking at
                ind1 = 0
            elif seq1[i-1] == 'C':
                ind1 = 1
            elif seq1[i-1] == 'G':
                ind1 = 2
            else:
                ind1 = 3
            if seq2[j-1] == 'A': #determine which element from sequence 2 we are looking at
                ind2 = 0
            elif seq2[j-1] == 'C':
                ind2 = 1
            elif seq2[j-1] == 'G':
                ind2 = 2
            else:
                ind2 = 3
            score_matrix[i][j] = score_matrix[i-1,j-1] + scoring_matrix[ind1][ind2] #add value from scoring_matrix to diag parent
            parent_matrix[i][j] = diag #set parent to diag
            
            if score_matrix[i-1][j] + gapPen > score_matrix[i][j]: #if gap up is larger than diag
                score_matrix[i][j] = score_matrix[i-1][j] + gapPen #add gap penalty and set parent to up
                parent_matrix[i][j] = up
            if score_matrix[i][j-1] + gapPen > score_matrix[i][j]: #if gap left is larger than above
                score_matrix[i][j] = score_matrix[i][j-1] + gapPen #add gap penalty and set parent to left
                parent_matrix[i][j] = left
            
                
                
    #print score_matrix
    traceX = len(seq1) #traceback indices
    traceY = len(seq2)
    #print parent_matrix
    optimum_alignment = [] #array to hold optimum alignment
    elem = parent_matrix[traceX][traceY] #get last element's parent
    score = score_matrix[traceX][traceY] #find score
    while(traceX != 0) and (traceY != 0): #until we reach (0,0)
        if elem == diag:
            optimum_alignment.append([seq1[traceX-1], seq2[traceY-1]]) #if diag, both sequences are favoured
            traceX -= 1
            traceY -= 1
        elif elem == up: #if up, gap in seq2
            optimum_alignment.append([seq1[traceX-1], '_'])
            traceX -= 1
        elif elem == left: #if left, gap in seq1
            optimum_alignment.append(['_', seq2[traceY-1]])
            traceY -= 1
        elem = parent_matrix[traceX][traceY] #move up
    optimum_alignment.reverse() #Array needs to be reversed, since it was built from back to front
    res = np.array(optimum_alignment) #convert to array so we can transpose
   
    return res.transpose(), score        #return vals, use transpose to display sequences under each other
        
    
    
    
    
```
